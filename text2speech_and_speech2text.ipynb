```python
from gtts import gTTS
import librosa
import soundfile as sf


def text2speech(text,file_name):
  language = 'en'
  
  myobj = gTTS(text=text, lang=language, slow=False)

  name = file_name
  print(name)
  myobj.save(name)
  x,_ = librosa.load(name, sr=16000)
  sf.write(name, x, 16000)
```

```python
mytext = "Let us start with a single transformation. I have chosen a random image from google and will perform a horizontal flip."
text2speech(mytext,'/content/audio1.wav')
```

```python
import speech_recognition as sr
# initialize the recognizer
def speech2text(path):
  r = sr.Recognizer()

  with sr.AudioFile(path) as source:
      # listen for the data (load audio to memory)
      audio_data = r.record(source)
      # recognize (convert from speech to text)
      text = r.recognize_google(audio_data)
      print(text)
```

```python
speech2text('/content/audio1.wav')
```
